## 下載檔案

```python
import requests
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

response[:1000]
```

```python
r.content
```

```python
r.status_code
```

```python
import json
data = json.loads(response)

print(len(data))
data
```

```python
for d in data:
    print(d['title'])
```
