1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
print(len(r.text))
r.text
```

```python
data = json.loads(r.text)
print(type(data))
print(data)
```

```python
#小練習
testdic = {"a":3,"d":0}
print('vakues',[testdic[i] for i in testdic])
print('keys',[i for i in testdic])
```

```python
data1 = data[0]
print(type(data1))
#打印出 dictionary data1 中之所有 keys
for i in data1:
    print(i,end = ",")
data1
```

```python
print("資料筆數",len(data))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
# title createdAt commentCount likeCount
for datas in data:
    listt = []
    listt.append(datas['title'])
    listt.append(datas['createdAt'])
    listt.append(datas['commentCount'])
    listt.append(datas['likeCount'])
    print(listt[:])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數commentCount」與「平均按讚人數likeCount」
from statistics import mean

popular_url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=true'
unpopular_url = 'https://www.dcard.tw/_api/forums/pet/posts?popular=false'

popular = requests.get(popular_url)
unpopular = requests.get(unpopular_url)

pop_art = json.loads(popular.text)
unpop_art = json.loads(unpopular.text)

pop_comment_mean = mean([i["commentCount"] for i in pop_art])
pop_likeCount_mean = mean([i['likeCount'] for i in pop_art])

unpop_comment_mean = mean([i["commentCount"] for i in unpop_art])
unpop_likeCount_mean = mean([i['likeCount'] for i in unpop_art])

print(" 熱門文章\n","平均留言數",round(pop_comment_mean,1),"平均按讚數",round(pop_likeCount_mean,1))
print(' ----------')
print(" 非熱門文章\n","平均留言數",round(unpop_comment_mean,1),"平均按讚數",round(unpop_likeCount_mean,1))
```
