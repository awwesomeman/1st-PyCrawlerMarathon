## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find('div',class_="part_list_2").find_all('h3'):
    print(d.find('span', class_="date").text, d.find('a').text)
```

## 2. 取出現在時間兩小時內的新聞

```python
#取得今天時間並轉呈timestamp 

from datetime import datetime, timedelta

datetime.today() #datetime.now()
```

```python
#傳換成和網站上一致的時間格式

# datetime.timedelta(days=0, seconds=0, microseconds=0, milliseconds=0, minutes=0, hours=0, weeks=0)

current_time = datetime.strptime(datetime.today().strftime('%Y-%m-%d %H:%M'),'%Y-%m-%d %H:%M')
target_time = current_time - timedelta(hours=2) #datetime才能作加減，string不行
target_time #回朔兩小時前之時間
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

browser = webdriver.Chrome(executable_path='..\chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")


for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
html_source = browser.page_source
time.sleep(5)
#browser.quit();
```

```python
soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find('div',class_="part_list_2").find_all('h3'):
    
    time = datetime.strptime(d.find('span', class_="date").text,'%Y/%m/%d %H:%M')
    if time > target_time:
        print(d.find('span', class_="date").text, d.find('a').text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import datetime, timedelta

target_time = datetime.today() - timedelta(days=3)

target_year = target_time.strftime('%Y') # current_year為string
target_month = target_time.strftime('%m')
target_day = target_time.strftime('%d')

target_time1 = datetime.strptime(f'{target_year}-{target_month}-{target_day} 15:00','%Y-%m-%d %H:%M')#轉化目標日期為datetime
target_time2 = datetime.strptime(f'{target_year}-{target_month}-{target_day} 17:00','%Y-%m-%d %H:%M')
```

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time


browser = webdriver.Chrome(executable_path='..\chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

selectDay = Select(browser.find_element_by_id("selD"))
selectDay.select_by_value(target_day)
browser.find_element_by_id('button').click()

for i in range(30):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 100000);")
    
html_source = browser.page_source

soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find('div',class_="part_list_2").find_all('h3'):
    
    time = datetime.strptime(d.find('span', class_="date").text,'%Y/%m/%d %H:%M')
    if time >= target_time1 and time<=target_time2:
        print(d.find('span', class_="date").text, d.find('a').text)
```
